```python
import mnist_loader
from network import Network, Sigmoid, CrossEntropyCost
```

# Load the MNIST data

```python
training_data, validation_data, test_data = mnist_loader.load_data_wrapper()
```

# Initialize the network

```python
net = Network([784, 50, 10], cost=CrossEntropyCost, activation=Sigmoid)
```

# Train

```python
net.SGD(training_data, epochs=25, mini_batch_size=10, eta=0.05, lmbda=5.0, evaluation_data=validation_data)
```

# Plot Performance

```python
net.plot_performance()
```
